```python
from scipy.signal import butter, filtfilt

# Bandpass filter function

def bandpass_filter(data, lowcut, highcut, fs, order=4):
    nyquist = 0.5 * fs  # Nyquist frequency

    # Auto-adjust invalid values
    if highcut >= nyquist:
        highcut = 0.9 * nyquist  # Set highcut to 90% of Nyquist
    if lowcut <= 0:
        lowcut = 0.1  # Avoid zero or negative frequencies

    low = lowcut / nyquist
    high = highcut / nyquist

    # Design and apply filter
    b, a = butter(order, [low, high], btype="band")
    return filtfilt(b, a, data)

# Define frequency ranges for each signal type
signal_filters = {
    "EEG": (0.5, 40),
    "EOG": (0.3, 35),
    "EMG": (10, 100),
    "ECG": (0.5, 50),
    "Resp": (0.1, 0.5)  # Adjust as needed
}

# Process each PSG file
filtered_signals = {}

for i in range(len(psg_files)):
    print(f"\nFiltering File: {psg_files[i]}")

    # Load PSG file
    raw = mne.io.read_raw_edf(psg_files[i], preload=True)
    sfreq = raw.info['sfreq']  # Sampling frequency
    signal_names = raw.ch_names  # Available signals

    filtered_data = {}

    # Apply filters to each signal based on type
    for signal in signal_names:
        raw_signal = raw.get_data(picks=[signal])[0]

        # Find the correct filter (default to EEG if unknown)
        filter_range = signal_filters.get(signal.split()[0], (0.5, 40))

        # Apply bandpass filter
        filtered_signal = bandpass_filter(raw_signal, filter_range[0], filter_range[1], sfreq)
        filtered_data[signal] = filtered_signal

    # Store filtered signals for further processing
    filtered_signals[psg_files[i]] = filtered_data
    print(f"Filtered {len(filtered_data)} signals from {psg_files[i]}")
```

```python
import numpy as np
import mne

def segment_psg_and_labels(psg_file, hypnogram_file, epoch_duration=30):
    """Segments all physiological signals into 30-second epochs aligned with sleep stage labels."""

    # Load PSG signals
    raw = mne.io.read_raw_edf(psg_file, preload=True)
    signals = raw.get_data()  # Shape: (n_channels, n_samples)
    fs = int(raw.info['sfreq'])  # Sampling frequency
    samples_per_epoch = fs * epoch_duration  # Number of samples per 30s epoch

    # Load hypnogram (sleep stage labels)
    annot = mne.read_annotations(hypnogram_file)

    # Define sleep stage mapping
    sleep_stage_mapping = {
        "Sleep stage W": 0,  # Wake
        "Sleep stage 1": 1,  # N1
        "Sleep stage 2": 2,  # N2
        "Sleep stage 3": 3,  # N3 (Deep Sleep)
        "Sleep stage 4": 3,  # N4 (Merged with N3)
        "Sleep stage R": 4   # REM Sleep
    }

    # Convert annotation descriptions to numerical labels using the mapping
    stage_labels = np.array([sleep_stage_mapping.get(desc, -1) for desc in annot.description])
    num_epochs = len(stage_labels)

    # Segment all signals into 30-second epochs
    segmented_signals = []
    for i in range(num_epochs):
        start_idx = i * samples_per_epoch
        end_idx = start_idx + samples_per_epoch
        if end_idx > signals.shape[1]:  # Avoid overflow
            break
        segmented_signals.append(signals[:, start_idx:end_idx])  # Store all channels

    segmented_signals = np.array(segmented_signals)  # Shape: (num_epochs, n_channels, samples_per_epoch)

    return segmented_signals, stage_labels
```

```python
from scipy.stats import skew, kurtosis
from scipy.signal import welch

def extract_features(segmented_signals, fs=100):
    """Extracts time-domain and frequency-domain features for each epoch."""
    feature_list = []

    for epoch in segmented_signals:
        epoch_features = []

        for signal in epoch:  # Process each physiological signal separately

            # Time-Domain Features
            mean_val = np.mean(signal)
            var_val = np.var(signal)
            skew_val = skew(signal)
            kurtosis_val = kurtosis(signal)
            rms_val = np.sqrt(np.mean(signal**2))
            peak_to_peak = np.ptp(signal)
            zero_crossings = np.sum(np.diff(np.sign(signal)) != 0)

            # Frequency-Domain Features (Power in different bands)
            freqs, psd = welch(signal, fs=fs, nperseg=fs*2)  # Compute PSD

            def bandpower(low, high):
                mask = (freqs >= low) & (freqs <= high)
                return np.trapz(psd[mask], freqs[mask])  # Integrate PSD in band

            delta_power = bandpower(0.5, 4)
            theta_power = bandpower(4, 8)
            alpha_power = bandpower(8, 12)
            beta_power = bandpower(12, 30)
            gamma_power = bandpower(30, 50)

            # Combine all features
            epoch_features.extend([
                mean_val, var_val, skew_val, kurtosis_val, rms_val,
                peak_to_peak, zero_crossings, delta_power, theta_power,
                alpha_power, beta_power, gamma_power
            ])

        feature_list.append(epoch_features)

    return np.array(feature_list)

# Run feature extraction
features = extract_features(segmented_signals)
print(f"Extracted Features Shape: {features.shape}")  # Expected: (num_epochs, num_features)
```

```python
!pip install antropy
import numpy as np
import pandas as pd
import mne
from scipy.stats import skew, kurtosis
from scipy.signal import welch
import antropy as ant

FREQ_BANDS = {
    "delta": (0.5, 4),
    "theta": (4, 8),
    "alpha": (8, 13),
    "beta": (13, 30),
    "gamma": (30, 45)
}

def compute_band_power(signal, sfreq, band):
    """Compute the power spectral density for a given frequency band."""
    fmin, fmax = FREQ_BANDS[band]
    f, Pxx = welch(signal, sfreq, nperseg=2*sfreq)
    return np.sum(Pxx[(f >= fmin) & (f <= fmax)])

def extract_features(raw, epochs, labels):
    """Extract features from EEG, EOG, and EMG signals."""
    sfreq = raw.info['sfreq']
    feature_data = []
    ch_names = raw.ch_names

    num_epochs = min(len(epochs.events), len(labels))

    epochs_data = epochs.get_data()

    for epoch_idx in range(num_epochs):
        epoch = epochs_data[epoch_idx]
        features = {}

        for ch_idx, ch_name in enumerate(ch_names):
            signal = epoch[ch_idx, :]

            # Time-domain features
            features[f"{ch_name}_mean"] = np.mean(signal)
            features[f"{ch_name}_variance"] = np.var(signal)
            features[f"{ch_name}_skewness"] = skew(signal)
            features[f"{ch_name}_kurtosis"] = kurtosis(signal)

            # Frequency-domain features
            for band in FREQ_BANDS.keys():
                features[f"{ch_name}_{band}_power"] = compute_band_power(signal, sfreq, band)

            # Nonlinear features
            features[f"{ch_name}_sample_entropy"] = ant.sample_entropy(signal)
            features[f"{ch_name}_higuchi_fd"] = ant.higuchi_fd(signal)

        features["Sleep Stage"] = labels[epoch_idx]
        feature_data.append(features)

    return pd.DataFrame(feature_data)

feature_df = extract_features(raw, epochs, labels)
print(feature_df.head())

feature_df.to_csv("sleep_features.csv", index=False)
```

```python
# STEP 1: Function to Compute Sleep Score Based on Extracted Metrics
def compute_sleep_score(labels, epoch_length=30, total_time_in_bed=8):
    """
    Compute sleep score based on extracted sleep stages.

    Parameters:
        labels (array): Array of sleep stage labels.
        epoch_length (int): Duration of each epoch in seconds.
        total_time_in_bed (int): Assumed total time in bed (hours).

    Returns:
        sleep_score (float): Calculated sleep score (0-100 scale).
        sleep_metrics (dict): Dictionary containing computed sleep metrics.
    """
    total_epochs = len(labels)
    total_sleep_epochs = np.sum(labels != 0)  # Count epochs that are not wake stage

    TST = (total_sleep_epochs * epoch_length) / 3600  # Total Sleep Time (hrs)
    Sleep_Efficiency = (TST / total_time_in_bed) * 100

    deep_sleep_epochs = np.sum(labels == 3)  # Deep Sleep (Stage N3)
    rem_sleep_epochs = np.sum(labels == 4)  # REM Sleep

    if total_sleep_epochs == 0:
        deep_sleep_percentage = 0
        rem_sleep_percentage = 0
    else:
        deep_sleep_percentage = max((deep_sleep_epochs / total_sleep_epochs) * 100, 5)
        rem_sleep_percentage = max((rem_sleep_epochs / total_sleep_epochs) * 100, 5)


    # STEP 2: Define Sleep Score Formula
    sleep_score = (0.5 * Sleep_Efficiency) + (0.4 * deep_sleep_percentage) + (0.2 * rem_sleep_percentage)


    sleep_metrics = {
        "Total Sleep Time (hrs)": TST,
        "Sleep Efficiency (%)": Sleep_Efficiency,
        "Deep Sleep (%)": deep_sleep_percentage,
        "REM Sleep (%)": rem_sleep_percentage,
        "Sleep Score": sleep_score
    }

    return sleep_score, sleep_metrics

```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# Reshape input data
X_train = np.reshape(X_train, (X_train.shape[0], 55, 1))
X_test = np.reshape(X_test, (X_test.shape[0], 55, 1))

# Define Pure LSTM Model
model = Sequential([
    LSTM(128, return_sequences=True, input_shape=(55, 1)),
    LSTM(64, return_sequences=True),
    LSTM(32, return_sequences=False),

    Dropout(0.3),
    Dense(32, activation='relu'),
    Dense(5, activation='softmax')  # 5 Sleep Stages
])

# Compile Model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train Model
history = model.fit(
    X_train, y_train,
    epochs=200, batch_size=16,
    validation_data=(X_test, y_test),
    verbose=1
)

# Evaluate Model
test_loss, test_acc = model.evaluate(X_test, y_test)
print(f"\n✅ Test Accuracy: {test_acc:.4f}")
```
